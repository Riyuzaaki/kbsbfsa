```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("Train_data.csv")
test=pd.read_csv("Test_data.csv")
```

```python
train.drop(['accident'],axis=1,inplace=True)
```

```python
train
```

```python
test
```

```python
vertical_stack = pd.concat([train, test], axis=0)
```

```python
vertical_stack
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
code=le.fit_transform(vertical_stack['code'].values)
code
```

```python
clientType=le.fit_transform(vertical_stack['clientType'].values)
clientType
```

```python
registrationMode=le.fit_transform(vertical_stack['registrationMode'].values)
registrationMode
```

```python
planName=le.fit_transform(vertical_stack['planName'].values)
planName
```

```python
country=le.fit_transform(vertical_stack['country'].values)
country
```

```python
x=pd.DataFrame()
x['code']=code
x['clientType']=clientType
x['registrationMode']=registrationMode
x['planName']=planName
x['country']=country
```

```python
x
```

```python
x['duration']=np.array(vertical_stack['duration'])
x['age']=np.array(vertical_stack['age'])
```

```python
x
```

```python
x_train=x[:60326]
```

```python
x_test=x[60326:]
```

```python
ykeliy=pd.read_csv("Train_data.csv")
ykeliy
```

```python
y_train=ykeliy['accident']
```

```python
x_train.isnull().sum()
```

```python
y_train.isnull().sum()
```

```python
x_test.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100)
```

```python
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
result=pd.DataFrame()
result['accident']=y_pred
```

```python
result.to_csv('result.txt', sep='\n', index=False, header=False)
```
